```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sig = np.linspace(-15,15,200)
```

```python
y0 = sigmoid(sig)
y1 = sigmoid(sig+8)
y2 = sigmoid(sig-8)
```

```python
plt.plot(sig,y0)
plt.plot(sig,y1)
plt.plot(sig,y2)
plt.xlabel('X')
plt.ylabel('Y')
```
